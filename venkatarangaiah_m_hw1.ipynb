Now let us define our functions for gradient descent 

```python
def sigmoid(z):
    return 1.0 / (1 + np.exp(-z))

def hypo(x,theta):
    z = np.dot(x,theta.T)
    ynew = sigmoid(z)
    return ynew 

def cost_function(x,y,theta):
    ynew = hypo(x,theta)
    cost = np.dot(y.T, np.log(ynew)) + np.dot((1-y).T, np.log(1-ynew))
    cost = -cost/x.shape[0]
    return cost

def log_function(theta, X):
    X=sf[:,:-1]
    z = np.dot(X,theta)
    return  1.0/( 1.0 + np.exp(-z))

def cross_entropy(theta, X, y):
    hx = log_function(X, theta)
    c = -y.dot(np.log(hx)) - ((1 - y).dot(np.log(1-hx)))
    return c

def gradient(x,y,theta):
    ynew = hypo(x,theta)
    grad = (np.dot(x.T, y-ynew))/700
    return grad

def gradient_descent(Train_X,Train_Y,Test_X,Test_Y, earlystopX, earlystopy, alpha, itiration):
    n = Train_X.shape[1]
    theta = np.zeros(n)
    cost = cost_function(Train_X, Train_Y, theta)
    cost_result = [cost]
    cost = cost_function(earlystopX, earlystopy, theta)
    cost_earlystop = [cost]
    
    for i in range(itiration):
        grad = gradient(Train_X, Train_Y, theta)
        theta = theta + alpha*grad
        cost_cal = cost_function(Train_X, Train_Y, theta)
        cost_result.append(cost_cal)
        cost_cal = cost_function(earlystopX, earlystopy, theta)
        cost_earlystop.append(cost_cal)
    return cost_result, cost_earlystop
```
